```python
import numpy as np
import pandas as pd
from scipy.spatial import distance_matrix

import matplotlib.pyplot as plt


#https://iugspace.iugaza.edu.ps/bitstream/handle/20.500.12358/24630/Optimization_Techniques_for_Solving_Travelling_Sal.pdf
```

```python
# BERLIN 52 DATASET
data = pd.read_csv("https://raw.githubusercontent.com/pdrozdowski/TSPLib.Net/master/TSPLIB95/tsp/berlin52.tsp", engine="python", delim_whitespace=True, skiprows=6, skipfooter=2, header=None, names=["index", "X", "Y"])[["X", "Y"]]
np_data = data.to_numpy()
dist_matrix = distance_matrix(np_data, np_data, p=2)

data_sol = pd.read_csv("https://raw.githubusercontent.com/pdrozdowski/TSPLib.Net/master/TSPLIB95/tsp/berlin52.opt.tour", engine="python", skiprows=4, skipfooter=1, header=None, names=["index"])
data_sol["index"].iloc[-1] = data_sol["index"][0]
optimal_solution = (data_sol-1).to_numpy().flatten().tolist()
```

```python
# CH 150 DATASET
data = pd.read_csv("https://raw.githubusercontent.com/pdrozdowski/TSPLib.Net/master/TSPLIB95/tsp/ch150.tsp", engine="python", delim_whitespace=True, skiprows=6, skipfooter=1, header=None, names=["index", "X", "Y"])[["X", "Y"]]
np_data = data.to_numpy()
dist_matrix = distance_matrix(np_data, np_data, p=2)

data_sol = pd.read_csv("https://raw.githubusercontent.com/pdrozdowski/TSPLib.Net/master/TSPLIB95/tsp/ch150.opt.tour", engine="python", skiprows=5, skipfooter=1, header=None, names=["index"])
data_sol["index"].iloc[-1] = data_sol["index"][0]
optimal_solution = (data_sol-1).to_numpy().flatten().tolist()
```

```python
def tour_size(dist_matrix, solution):
    tour = 0
    prev_idx = solution[0]
    for idx in solution[1:]:
        tour += dist_matrix[prev_idx, idx]
        prev_idx = idx
    tour += dist_matrix[prev_idx, 0]
    return tour
```

```python
def plot_solution(data, solution=[], optimal=None):
    plt.figure(figsize=(12,7))
    plt.scatter(data.X, data.Y)
    plt.plot(data.X[solution], data.Y[solution])
    if optimal is not None:
        plt.plot(data.X[optimal], data.Y[optimal])
```

```python
def solve_bottom_up(dist_matrix):
    solution = [0,1,2]

    for city_idx in range(3, dist_matrix.shape[0]):
        best = np.infty
        
        for j in range(len(solution)):
            prev_city_idx = solution[j]
            next_city_idx = solution[(j+1) % len(solution)]

            additional = dist_matrix[prev_city_idx, city_idx] + dist_matrix[next_city_idx, city_idx] - dist_matrix[prev_city_idx, next_city_idx]
            if additional < best:
                insert_idx = (j+1) % len(solution)
                best = additional

        solution.insert(insert_idx, city_idx)
    solution.append(solution[0])
    return solution
```

```python
plot_solution(data)
```

```python
solution = solve_bottom_up(dist_matrix)
print("Found: ", tour_size(dist_matrix, solution))
print("Optimal: ", tour_size(dist_matrix, optimal_solution))
```

```python
plot_solution(data, solution, optimal=optimal_solution)
```
